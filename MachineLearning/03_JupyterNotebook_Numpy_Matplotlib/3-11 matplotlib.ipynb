## matplotlib 基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 10, 100)
x
```

```python
y = np.sin(x)
y
```

```python
plt.plot(x, y)
```

```python
siny = y.copy()
```

```python
cosy = np.cos(x)
cosy.shape
```

```python
plt.plot(x, siny)
# plt.plot(x, cosy, color='red') # 可以指定颜色
plt.plot(x, cosy, '--g')
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlim(-1, 11)  # 设定x轴范围
plt.ylim(-2, 2)  # 设定y轴范围
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.axis([-1, 11, -2, 2])  # 同时设置x，y轴范围， 前两个表示x范围，后两个表示y范围
```

```python
plt.plot(x, siny)
plt.plot(x, cosy)
plt.xlabel('x axis')
plt.ylabel('y value')
```

```python
plt.plot(x, siny, label='sin(x)')
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis')
plt.ylabel('y value')
plt.legend()
```

```python
plt.plot(x, siny, label='sin(x)')  # 加入图示
plt.plot(x, cosy, label='cos(x)')
plt.xlabel('x axis')  # x轴名字
plt.ylabel('y value')
plt.legend()  # 图示
plt.title('Welcome to the ML world')  # 标题
```
